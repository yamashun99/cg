```python
import numpy as np
```

```python
N = 100
A = np.random.rand(N, N)
A = A @ A.T
b = np.random.rand(N)
```

```python
# x = np.random.rand(N)
x = np.zeros(N)
r = b - A @ x
```

```python
for i in range(100000):
    alpha = (r @ r) / (r @ A @ r)
    x = x + alpha * r
    r = b - A @ x
    print(np.linalg.norm(r))
```
