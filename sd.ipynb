```python
import numpy as np
```

```python
N = 10
A = np.zeros((N, N))
for i in range(N):
    for j in range(N):
        A[i, j] = N - abs(i - j)
print(A)
b = np.random.rand(N)
```

```python
x = np.random.rand(N)
r = b - A @ x
```

```python
for i in range(100000):
    alpha = (r @ r) / (r @ A @ r)
    x = x + alpha * r
    r = b - A @ x
    print(np.linalg.norm(r))
```
